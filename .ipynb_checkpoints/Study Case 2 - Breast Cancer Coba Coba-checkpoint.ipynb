```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import Normalizer, normalize
from sklearn.cross_validation import KFold, cross_val_score
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import f1_score, confusion_matrix
from sklearn.metrics import accuracy_score

bcdf = pd.read_csv("./dataset/data breast cancer.csv")
bcdf.shape
```

```python
# Analisa data
# 1. kolom id yang tidak bisa kita gunakan untuk klasifikasi
# 2. kolom diagnosis adalah kolom yang akan kita klasifikasi
# 3. kolom ke 32 terdapat 'Unnamed: 32' yang isinya NaN
```

```python
col = bcdf.columns
print (col)
```

```python
# Kolom yang diapus
bcdf = bcdf.drop(['Unnamed: 32','id','diagnosis'], axis=1)
```

```python
# Data Preprocessing

X = bcdf.iloc[:,:-1]   # 1. -1 karena untuk menghilangkan kolom 'id'
y = bcdf.diagnosis  # 2. kolom yang akan kita klasifikasi M dan B

# Use 80% data for training

train_size = 0.8

# Split data into testing and training set

X_train, X_test, y_train, y_test = train_test_split(X, y, train_size = train_size)
```

```python
# Diagnosis

ax = sns.countplot(y, label="Count")
B, M = y.value_counts()
print ("Number of Benign", B)
print ("Number of Malignant", M)
```

```python
f,ax = plt.subplots(figsize=(18,18))
sns.heatmap(X.corr(), annot=True, linewidths=.5, fmt='.lf', ax=ax)
```

```python
# Normalization
norm = Normalizer()

# Fit
norm.fit(X_train)

# Transform both training and testing sets
X_train_norm = norm.transform(X_train) 
X_test_norm = norm.transform(X_test)
```
